Вопрос 1: Сколькими способами можно выбрать 5 любых компаний для обзвона?

Ответ дайте в виде целого числа

````python
from math import factorial
````

````python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n-k)))
````
